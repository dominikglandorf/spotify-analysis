# Get data from Spotify Web API
This requires setting the environment variables CLIENT_ID and CLIENT_SECRET.

```python
import requests
import string
import time
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv("../")
```

# Requesting an oAuth2 access token
The Spotify Web API required OAuth2 authentication.

```python
def get_access_token(url, client_id, client_secret):
    response = requests.post(
        url,
        data={"grant_type": "client_credentials"},
        auth=(client_id, client_secret),
    )
    return response.json()["access_token"]

auth_url = "https://accounts.spotify.com/api/token"
client_id = os.environ['CLIENT_ID']
client_secret = os.environ['CLIENT_SECRET']
access_token = get_access_token(auth_url, client_id, client_secret)
print(access_token)
```

# Get artist data for Justin Bieber
He has the unique Spotify ID: 1uNFoZAHBGtllmzznpCI3s?si=O31o0DVURguEKhGhaodkyw

```python
def get_authenticated(url):
    return requests.get(url,
             headers={'Authorization': 'Bearer '+ access_token})
    

response = get_authenticated("https://api.spotify.com/v1/artists/1uNFoZAHBGtllmzznpCI3s")
artist = response.json()
print(f"Justin Bieber's popularity is {artist['popularity']}. He has {artist['followers']['total']} followers.")
```

# Get a list of artists
To analyse music of different artists, we need some artists, that's why we search for A to Z artists to get some.

```python
letters = list(string.ascii_lowercase)
artists = {}

for l in letters:
    print(l)
    
    response = get_authenticated(f"https://api.spotify.com/v1/search?q={l}&limit=50&type=artist")
    search = response.json()
    artists_found = [{'id': artist['id'], 'name': artist['name'], 'popularity': artist['popularity']} for artist in search['artists']['items']]
    for a in artists_found:
        if not a['id'] in artists:
            artists[a['id']] = a
    time.sleep(2)
```

```python
len(artists)
```

# Get tracks for the artists
Let's get the 10 first tracks for the artist.

```python
all_tracks = []

i = 0
for a in list(artists.items()):
    print(f"{i} {a[1]['name']}")
    i += 1
    response = get_authenticated(f"https://api.spotify.com/v1/artists/{a[0]}/top-tracks?market=DE")
    search = response.json()
    #print(search)
    if not 'tracks' in search: continue
    tracks = {t['id']: {'name': t['name'], 'artist': a[1]['name'], 'popularity': t['popularity']} for t in search['tracks']}
    
    ids_list = ",".join(tracks.keys())
    response = get_authenticated(f"https://api.spotify.com/v1/audio-features?ids={ids_list}")
    for f in response.json()['audio_features']:
        if f is not None:
            tracks[f['id']]['danceability'] = f['danceability']
            tracks[f['id']]['energy'] = f['energy']
            tracks[f['id']]['liveness'] = f['liveness']
            tracks[f['id']]['loudness'] = f['loudness']
            tracks[f['id']]['speechiness'] = f['speechiness']
            tracks[f['id']]['acousticness'] = f['acousticness']
            tracks[f['id']]['instrumentalness'] = f['instrumentalness']
            tracks[f['id']]['tempo'] = f['tempo']
            tracks[f['id']]['valence'] = f['valence']
            tracks[f['id']]['key'] = f['key']
            
    all_tracks = all_tracks + list(tracks.values())

    time.sleep(0.25)
```

```python
tracks_df = pd.DataFrame(all_tracks)
print(tracks_df.shape)
tracks_df.head()
```
